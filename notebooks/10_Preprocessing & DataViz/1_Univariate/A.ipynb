## Pivot Tables

### Transform object to DataFrame

> Only if you don't have a DataFrame yet.

```python
#Transformación a Data Frame.
df_solar = enerSolar.to_frame()
df_solar
```

### Create columns with datetime information

```python
#Creamos nuevas columnas en el Data Frame asignando los valores accesibles gracias al formato DateTime del índice.
df_solar_pv = (df_solar
 .assign(
     year = df_solar.index.year, #Método que obtiene el año.
     month = df_solar.index.month, #Método que obtiene el mes.
     day = df_solar.index.day, #Método que obtiene el dia.
     hour = df_solar.index.hour #Método que obtiene la hora.
    )
)
```

```python
#Comprobación de la creación de nuevas columnas en el Data Frame.
df_solar_pv
```

### Process `pivot_table`

- Use this methodology.

```python
uf_pivotTable = (df_solar_pv
    .pivot_table( #Función para crear una Tabla Pivot.
        values='Solar Photovoltaic', #Selección del conjunto de valores para la tabla.
        index='hour', #Selección del índice que tendra la tabla.
        columns=['year','month'], #Selección de la columna que tendra la tabla (una lista para pasar más de 1 columna).
        aggfunc=sum)) #Función agregada que realizará y mostrata la Tabla Pivot.
```

```python
#Impresión de la tabla Pivot sin estilizar.
uf_pivotTable
```

### Style DataFrame

```python
#Modificación y mejora visual de la pivot Table
final_pivotTable = (uf_pivotTable
 .loc[6:22,:] #Limita los registros y campos mostrados, aquí limitamos las horas de luz solar.
 .div(1_000) #Divide los datos numericos entre un número especificado, permite hacer escalas y mejorar su visualización.
 .style #Permite introducir parametros para mejorar la presentación de la tabla.
    .background_gradient('Oranges') #Coloca un color degradado en razón del valor más bajo al más alto.
    .format(precision=0) #Nos permite especificar la cantidad de decimales que se muestran.
)

#Fue requerida la instalación de Jinja2
#!pip install Jinja2
```

```python
final_pivotTable
```

## Resampling the Time
- [Date offset `rule` list](https://pandas.pydata.org/docs/user_guide/timeseries.html#dateoffset-objects)
- [Pandas plot `kind` list](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.html)

```python
#Utilizamos el método resample para segregar datos por algún periodo temporal (revisar documentación)
#Creamos un DF que almacena la suma diaria de todos los valores.
df_solar_daily = df_solar.resample(rule='D').sum()
```

```python
#Grafica sobre los valores diarios de la géneración de energia Foto Voltaica.
df_solar_daily.plot()
```

```python
df_solar_monthly = df_solar.resample(rule='M').sum()
```

```python
#Grafica sobre los valores mensuales de la géneración de energia Foto Voltaica.
df_solar_monthly.plot()
```

```python
df_solar_bimester = df_solar.resample(rule='2M').sum()
```

```python
df_solar_bimester.plot()
```

```python
df_solar_trimester = df_solar.resample(rule='3M').sum()
```

```python
df_solar_trimester.plot()
```
